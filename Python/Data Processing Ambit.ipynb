```python
import pandas as pd

# Read the CSV file and skip the first row
df = pd.read_csv('/content/drive/MyDrive/AUT_export_Tower.csv', skiprows=0)

# Drop columns that do not need to be read
df.drop(df.columns[[4]], axis=1, inplace=True)
df.head()
```

```python
import datetime

df = pd.DataFrame(df)

# Convert Timestamp column to datetime format
df["Timestamp"] = pd.to_datetime(df["Timestamp"])

# Sort the DataFrame by "Conversation ID" and "Timestamp"
sorted_df = df.sort_values(by=["Conversation ID", "Timestamp"])

# Print the sorted DataFrame
print(sorted_df)
sorted_df.head()
```
